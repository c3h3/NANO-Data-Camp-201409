```python
ContinuousTypes = [np.int_,  np.int8, np.int16, np.int16, np.int32, np.int64, 
                   np.float_, np.float16, np.float32, np.float64, np.float128]
```

```python
DiscreteTypes = [np.str_, np.str, np.unicode_, np.unicode]
```

```python
def auto_convert_to_orange_domains(column_type):
    if column_type in DiscreteTypes:
        return Orange.feature.Discrete
    elif column_type in ContinuousTypes:
        return Orange.feature.Continuous
    else:
        return Orange.feature.Python
```

```python
def convert_string_to_orange_column_type(type_string):
    assert type_string in ["Continuous", "Discrete", "Python"]
    if type_string == "Discrete":
        return Orange.feature.Discrete
    elif type_string == "Continuous":
        return Orange.feature.Continuous
    else:
        return Orange.feature.Python
```

```python
output_column = 'Survived'
skip_columns = []
column_types = {'Survived':"Discrete"}
problem_type = "classification"
discrete2string = 50
```

```python
def df_to_orange_table(df, output_column = None, skip_columns = [], column_types = {}, problem_type = "classification", discrete2string = 50):
    assert problem_type in ["classification", "regression"]
    assert isinstance(discrete2string, int)
    
    cols_list = [ one_col for one_col in df.columns if (one_col != output_column) and (not one_col in skip_columns)] 
    
    if output_column != None:
        cols_list =  cols_list + [output_column]
        
    domains_list = []
    data_list = []
    
    
    

        
        
```

```python
for u in column_types:
    print u
```

```python
_column_types = {}
for one_col in column_types:
    _column_types[one_col] = convert_string_to_orange_column_type(column_types[one_col])
```

```python
_column_types
```

```python
coldata = {}
for one_col in df.columns:
    one_col_arr = np.array(df[one_col].values.tolist())
    coldata[one_col] = {"type":one_col_arr.dtype.type, "data_array":one_col_arr}
    
```

```python
coltypes_dict = dict(zip(df.columns,df.apply(lambda xx:auto_convert_to_orange_domains(np.array(xx.values.tolist()).dtype.type)).tolist()))
```

```python
for one_col in column_types.keys():
    coltypes_dict[one_col] = convert_string_to_orange_column_type(column_types[one_col])
```

```python
if problem_type == "classification":
    coltypes_dict[output_column] = Orange.feature.Discrete
    if not coldata[output_column]["type"] in DiscreteTypes:
        coldata[output_column]["data_array"] = coldata[output_column]["data_array"].astype(DiscreteTypes[0])
        coldata[output_column]["type"] = DiscreteTypes[0]
else:
    coltypes_dict[output_column] = Orange.feature.Continuous
    
```

```python
coltypes_dict
```

```python

colvalues = {}

for one_col in df.columns:
    
    one_col_arr = coldata[one_col]["data_array"]
    
    
    if coltypes_dict[one_col] == Orange.feature.Discrete:
        if discrete2string > 0:
            print one_col
            unique_values = np.unique(one_col_arr)
            print len(unique_values)
            
            if len(unique_values) > discrete2string:
                coltypes_dict[one_col] = Orange.feature.String
            else:
                colvalues[one_col] = unique_values.tolist()
        
```

```python
colvalues
```

```python
pre_domains = []

for one_col in df.columns:
    if one_col != output_column:
        if one_col in colvalues.keys():
            pre_domains.append(coltypes_dict[one_col](one_col, values=colvalues[one_col]))
        else:
            pre_domains.append(coltypes_dict[one_col](one_col))
    
    
if output_column in colvalues.keys():
    pre_domains.append(coltypes_dict[output_column](output_column, values=colvalues[output_column]))
else:
    pre_domains.append(coltypes_dict[output_column](output_column))
        
```

```python
domains = Orange.data.Domain([coltypes_dict[one_col](one_col) for one_col in df.columns if one_col != output_column] + [coltypes_dict[output_column](output_column)] )
```
